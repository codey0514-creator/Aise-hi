N / 2^k = 1 (because after a k times the list becomes of 1 element)

by solving this k becomes log N (run time)

```python
def test_location(l1 , index , guess):
    if l1[index-1] != guess:
        return index
    else :
        return test_location(l1 , index - 1 , guess)
def locate_cards(l1, guess):
    low , high = 0 , len(l1) - 1
    while low <= high :
        middle = (low + high)//2
        if l1[middle] == guess :
            return middle
        elif l1[middle] > guess :
            low = middle + 1
        elif l1[middle] < guess :
            high = middle - 1
    return test_location(l1 , middle , guess)
l1 = list(range(10000000 , 0 , -1))
print(locate_cards(l1 , 2))
```

***Question. 2) Given an array of integers nums sorted in ascending order, find the starting and ending position of a number***

```python
def searchRange( nums, target: int) :
    if not nums :
        return [-1,-1]
    return [-1,-1] if first(nums,target) == [-1,-1] else [first(nums , target) , last(nums , target)]
def first(nums , target ):
    low , high = 0 , len(nums) - 1
    while low <= high:
        middle = (low + high) // 2
        if nums[middle] == target and middle != low:
            high = middle - 1
        elif nums[middle] == target and middle == low:
                return middle
        elif nums[middle] > target :
            high = middle - 1
        elif nums[middle] < target :
            low = middle + 1 
        if nums[middle - 1] != target  and nums[middle] == target:
            return middle
    return [-1,-1]
def last(nums , target ):
    low , high = 0 , len(nums) - 1
    while low <= high:
        middle = (low + high) // 2
        if nums[middle] == target and middle != low:
            low = middle + 1
        elif nums[middle] == target and middle == low :
            return high if nums[high] == target else low
        elif nums[middle] > target :
            high = middle - 1
        elif nums[middle] < target :
            low = middle + 1 
        if nums[middle + 1] != target  and nums[middle] == target:
            return middle
    return [-1,-1]
```

***Question 3.)***

**Given an array of unique elements it is found by rotating an array given in ascending order (moving the last element to the first) we need to find the minimum element of the array with complexity of O(log N) that is we need to do Binary Search**

```python
def find_element(nums):
    low , high = 0 , len(nums) - 1
    if not nums :
        return None
    while low <= high :
        middle = (low + high) // 2
        if nums[middle] < nums[high]:
            high = middle 
        elif nums[middle] > nums[high]:
            low = middle
        if high - low == 1 :
            return nums[low] if nums[low] < nums[high] else nums[high]
print(find_element([3,4,0,1,2]))
```

```python
def tree_to_tuple(data):
    if data is None :
        return None
    if data.left is None and data.right is None :
        return data.key
    else :
        return (tree_to_tuple(data.left) , data.key , tree_to_tuple(data.right))
```

# Define a function get_index that gives index of any string on the basis of given instruction
* Iterate over the string, character by character
* Convert each character to a number using Python's built-in ord function.
* Add the numbers for each character to obtain the hash for the entire string
* Take the remainder of the result with the size of the data list

```python
def get_index(data_list , a_string):
    result = 0
    for a_char in a_string :
        a_num = ord(a_char) 
        result += a_num
    index = result % len(data_list)
    return index
print(get_index(data_list , "Nishchal"))
```
